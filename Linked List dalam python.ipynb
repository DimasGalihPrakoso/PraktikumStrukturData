```python
#Hapus node di akhir
class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None

class LinkedList:
  def __init__(self):
    self.headval = None

#Print the linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print (printval.dataval)
      printval = printval.nextval

#Function to add in the end
  def AddBegining(self,newdata):
    NewNode = Node(newdata)
    if self.headval is None:
      self.headval = NewNode
      return
    last = self.headval
    while(last.nextval):
      last = last.nextval
    last.nextval=NewNode

#Function to remove last node
  def RemoveEnd(self):
    last = self.headval
    while (last is not None):
      if last.nextval == None:
        break
      prev = last
      last = last.nextval
    if (last == None):
      return
    prev.nextval = last.nextval
    last = None

#Create linkedlist named Li
Li = LinkedList()

#Add node
Li.AddBegining("Januari")
Li.AddBegining("Februari")
Li.AddBegining("Maret")
print("Sebelum dilakukan remove")
Li.listprint()
print("=============================================")

#Remove first node
Li.RemoveEnd()
print("Setelah dilakukan remove node akhir")
Li.listprint()
```

```python
#Tugas

#Double Linked List (3 node) + Sisip Tengah + Hapus Kata Kunci
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoubleLinkedList:
    def __init__(self):
        self.head = None

    def print_list(self):
        temp = self.head
        while temp:
            print(temp.data, end="-")
            temp = temp.next

#Membuat 3 node
dll = DoubleLinkedList()

n1 = Node("A")
n2 = Node("B")
n3 = Node("C")

dll.head = n1
n1.next = n2
n2.prev = n1
n2.next = n3
n3.prev = n2

#Sisip Tengah (antara B dan C)
new_node = Node("X")
new_node.next = n3
new_node.prev = n2
n2.next = new_node
n3.prev = new_node

#Hapus dengan kata kunci "B"
temp = dll.head
while temp:
    if temp.data == "B":
        if temp.prev:
            temp.prev.next = temp.next
        if temp.next:
            temp.next.prev = temp.prev
        break
    temp = temp.next

dll.print_list()
```

```python
#Circular Linked List (3 node) + Sisip Akhir + Hapus Akhir
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def print_list(self):
        temp = self.head
        if temp is None:
            return
        while True:
            print(temp.data, end=" > ")
            temp = temp.next
            if temp == self.head:
                break
        print("(Back to Head)")

cll = CircularLinkedList()

n1 = Node("Jumat")
n2 = Node("Sabtu")
n3 = Node("Minggu")

cll.head = n1
n1.next = n2
n2.next = n3
n3.next = cll.head

#Sisip akhir
new_node = Node("Senin")
temp = cll.head
while temp.next != cll.head:
    temp = temp.next
temp.next = new_node
new_node.next = cll.head

#Hapus akhir
prev = None
temp = cll.head
while temp.next != cll.head:
    prev = temp
    temp = temp.next
prev.next = cll.head

cll.print_list()
```
